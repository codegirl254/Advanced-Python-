```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("DataFiles/data.xlsx")
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"]) #convert to datetime formart

plt.plot(df["Date"],df["Sales"] ,color ="Green")
plt.title("Line graph")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.bar(df["Product"],df["Sales"] ,color ="purple",alpha = 0.1)
plt.title("Barchart")
plt.xlabel("Sales")
plt.ylabel("Sales")
plt.show()
```

```python
sales_data = df.groupby("Category")["Sales"].sum()
sales_data
```

```python
plt.figure(figsize=(8,4))
plt.pie(sales_data,labels = sales_data.index, autopct = "%1.1f%%")
plt.title("Propotion of Sales by Category")
plt.show()
```

```python
plt.hist(df["Sales"] ,bins=15, color ="blue",edgecolor ="white")
plt.title("Histogram")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.fill_between(df["Date"],df["Sales"] ,color ="lightgreen",alpha=0.5 )
plt.plot(df["Date"],df["Sales"] ,color ="Green")
plt.title("Area plot")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation = 45)
plt.show()
```
